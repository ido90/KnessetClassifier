```python
import threading
global tidx
global results

tidx = 0
results = [None]*len(topic_list)

class Runner(threading.Thread):
    def __init__(self, lock):
        self.lock = lock
        threading.Thread.__init__(self)
        
    def run(self):
        global tidx

        while tidx <= len(topic_list):
            self.lock.acquire()
            run_on = tidx
            tidx += 1
            print("Increasing to " + str(tidx))
            self.lock.release()

            res = lemmatize(topic_list[run_on])
            self.lock.acquire()
            global results
            results[run_on] = res
            self.lock.release()
             
```
